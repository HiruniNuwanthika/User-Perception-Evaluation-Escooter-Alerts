```python
import pandas as pd
import datetime
```

```python
def calculate_time_bracket(infile, outfile):
    # Function to convert min:sec to datetime.timedelta
    def convert_to_time(timestr):
        try:
            min, sec = map(int, timestr.split(':'))
            return datetime.timedelta(minutes=min, seconds=sec)
        except ValueError:
            return None 

    # Function to convert datetime.timedelta back to min:sec
    def timedelta_to_str(timedelta):
        total_seconds = int(timedelta.total_seconds())
        minutes, seconds = divmod(total_seconds, 60)
        return f"{minutes}:{seconds:02d}"

    df = pd.read_csv(infile)

    # Convert "time" column to timedelta
    df['en_time'] = df['en_time'].apply(convert_to_time)

    # Subtract 3 seconds and add 3 seconds
    df['time_minus_3s'] = df['en_time'].apply(lambda x: x - datetime.timedelta(seconds=3))
    df['time_plus_3s'] = df['en_time'].apply(lambda x: x + datetime.timedelta(seconds=3))

    # Convert timedelta back to min:sec format
    df['time_minus_3s'] = df['time_minus_3s'].apply(timedelta_to_str)
    df['time_plus_3s'] = df['time_plus_3s'].apply(timedelta_to_str)

    # Print the DataFrame to check results
    df.to_csv(outfile, index=False)

```

```python
infile=r'Fixations\WithRecordingTime\participant#_encounter_times.csv' #manually created file
outfile=r'Fixations\WithRecordingTime\participant#_encounter_times_calculated.csv'
calculate_time_bracket(infile, outfile)
```

```python
def convert_to_time(timestr):
    min, sec = map(int, timestr.split(':'))
    return datetime.timedelta(minutes=min, seconds=sec)

# Read CSV file A
df_a = pd.read_csv(r'Fixations\WithRecordingTime\participant#_encounter_times_calculated.csv')
df_a['time_minus_3s'] = df_a['time_minus_3s'].apply(convert_to_time)
df_a['time_plus_3s'] = df_a['time_plus_3s'].apply(convert_to_time)

# Read CSV file B
df_b = pd.read_csv(r'Fixations\WithRecordingTime\participant#.csv')
df_b['recording_time'] = df_b['recording_time'].apply(convert_to_time)
# Assuming df_b has a column named 'time' in min:sec format that we need to compare against
output_file = r'Fixations\WithRecordingTime\participant#_output.csv'
open(output_file, 'w').close()  # This clears the existing file contents

# Iterate over rows in df_a
for index, row in df_a.iterrows():
    # Filter rows in df_b based on 'minus' and 'plus' times from df_a
    mask = (df_b['recording_time'] >= row['time_minus_3s']) & (df_b['recording_time'] <= row['time_plus_3s'])
    filtered_df_b = df_b[mask]
    filtered_df_b['participant no'] = row['participant no']
    filtered_df_b['way'] = row['way']
    filtered_df_b['alert'] = row['alert']
    filtered_df_b.to_csv(output_file, mode='a', header=not index, index=False)
```
